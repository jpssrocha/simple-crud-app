# Serializing indexes

I've been having a problem with Postgres to input my old SQLite database as seed. It looks like something to do with the automatic id creation. I will solve it on python by serializing the link between my tables, and them removing the columns that have id that are automatically created;

```python
import pandas as pd
```

```python
# Load tables with unordered indexes

livros = pd.read_csv("bd_livraria-livros.csv")
publicacao = pd.read_csv("bd_livraria-publicacao.csv")

# Create dict to translate from unordered index to ordered index
dict_ = { id_:i for i, id_ in enumerate(livros.id, start=1)}

# Apply to primary table
livros.id = livros.id.apply(lambda x: dict_[x])
# Them to related table
publicacao.livro_id = publicacao.livro_id.apply(lambda x: dict_[x])

# Them drop id column, to not confuse the DBMS 
# since it will be automatically generated by him

livros = livros.drop(columns="id")
```

```python
livros.to_csv("bd_livraria-livros-serialized.csv", index=False)
publicacao.to_csv("bd_livraria-publicacao-serialized.csv", index=False)
```
